## Parameter space for the mutation gain/ loss rates
This was the first attempt where I considered the parameter space for the mutation rates and repeated that for different values of $\alpha$.

From this we learned that $\mu_{FM}$ was not a determinant parameter, and we should focus on the parameter space in terms of $\alpha$ and $\mu_{MF}$

```python
# Dividing the intervals into 10^3 evenly spaced intervals
muFM = np.arange(1e-8, 1e-6, (1e-6 - 1e-8) / 1e2)
muMF = np.arange(1e-3, 2e-1, (2e-1 - 1e-3) / 1e2)

X, Y = np.meshgrid(muMF, muFM)
positions = np.dstack([X.ravel(), Y.ravel()])[0]

# Make an array with the boundaries of each cell in the grid, it is neccessary to sample from the uniform distribution
muFM_interval = np.column_stack((muFM[:-1], muFM[1:]))
muMF_interval = np.column_stack((muMF[:-1], muMF[1:]))

## Sampling from a uniform distribution over each cell in the grid. The sampled points are different for each cell

# mu_MF [:,:,0], mu_FM [:,:,1]
sample_mu = np.zeros((muMF_interval.shape[0], muFM_interval.shape[0], 2))
for i0, i1 in enumerate(muMF_interval):
    for j0, j1 in enumerate(muFM_interval):
        temp_muMF = np.random.uniform(i1[0], i1[1])
        temp_muFM = np.random.uniform(j1[0], j1[1])
        sample_mu[i0, j0, 0] = temp_muMF
        sample_mu[i0, j0, 1] = temp_muFM
```

```python
def verify_convergence(x1 , x2, epsilon = 1e-4):
    '''
        This function calculates the convergence in the populations of two given days
        x1, x2: arrays of shape (day_duration, # of strains)
        epsilon: tolerance for estabilshing convergence, by default 10^-4

        Returns true if change in populations is less than threshold
    '''
    area1 = np.array([simpson(x1[:, 0]), simpson(x1[:, 1])])
    area2 = np.array([simpson(x2[:, 0]), simpson(x2[:, 1])])
    norm = np.linalg.norm(area1 - area2)

    return norm < epsilon   
```

# Parameter Space Exploration using the analytical solution

Considering the equations without the carrying cpacity term we can actually obtain an analytical solution. First let's define the parameters
$$a = \left(1-\mu_{FM} \right ) + \alpha\left(1-\mu_{MF} \right ) \\
b = \left(1-\mu_{FM} \right ) - \alpha\left(1-\mu_{MF} \right ) \\
\Delta =\left[  \left(1-\mu_{FM} \right ) + \alpha\left(1-\mu_{MF} \right )\right ]^2 - 4\alpha\left(1 - \mu_{FM} - \mu_{MF}\right)\\
\Delta = a^2+ 4\alpha\left(1 - \mu_{FM} - \mu_{MF}\right)\\
$$

In terms of these parameters the analytical solution is

$$
F(t)= \frac{e^{\frac{a}{2}t}}{\sqrt{\Delta}}\left[  b\sinh\left(\frac{\sqrt{\Delta}}{2}t \right ) + \sqrt{\Delta}\cosh\left(\frac{\sqrt{\Delta}}{2}t \right )\right ]\\ 
C(t) = \frac{2\mu_{FM}e^{\frac{a}{2}t}}{\sqrt{\Delta}}\sinh\left(\frac{\sqrt{\Delta}}{2}t \right )
$$

```python
def population(mu_FM, mu_MF, alpha, t):
    '''
        This function implements the analytical solution for the equations
        It need as input values for the replication rates and the mutation rates, as well as a time interval
        The time interval must be interpreted as time steps in units of r_F

        Returns two arrays, Founder, Mutant populations
    '''
    a = (1 - mu_FM) + alpha * (1 - mu_MF)
    b = (1 - mu_FM) - alpha * (1 - mu_MF)
    delta = a ** 2 - 4 * alpha * (1 - mu_FM - mu_MF)

    sqrt_delta = 0.5 * np.sqrt(delta)
    F = 2 * np.exp(0.5 * a * t) / sqrt_delta * (b * np.sinh(sqrt_delta * t) + 2 * sqrt_delta * np.cosh(sqrt_delta * t ))
    C = mu_FM * np.exp(0.5 * a * t) / sqrt_delta * np.sinh(sqrt_delta * t)

    return F,C
```

The total population as a function of time is given by
$$
T(t)=\frac{e^{\frac{a}{2}t}}{\sqrt{\Delta}}\left[\left(b + 2\mu_{FM}\right)\sinh{\left(\frac{\sqrt{\Delta}}{2}t\right)} + \sqrt{\Delta}\cosh{\left(\frac{\sqrt{\Delta}}{2}t\right)}\right] \\
$$
Therefore, the proportions for each strain are
$$
f(t) = \frac{b\tanh{{\left(\frac{\sqrt{\Delta}}{2}t\right)}} + \sqrt{\Delta}}{\left(b + 2\mu_{FM}\right)\tanh{\left(\frac{\sqrt{\Delta}}{2}t\right)} + \sqrt{\Delta}}\\
m(t) = \frac{2\mu_{FM}}{b + 2\mu_{FM} + \sqrt{\Delta}\coth{\left(\frac{\sqrt{\Delta}}{2}t\right)}} 
$$

In the limit $t\rightarrow \infty$
$$
f_\infty = \frac{b + \sqrt{\Delta}}{b + 2\mu_{FM} + \sqrt{\Delta}}\\
\ \\
m_\infty = \frac{2\mu_{FM}}{b + 2\mu_{FM} + \sqrt{\Delta}}
$$

```python
@np.vectorize
def population_limit(mu_FM, alpha, mu_MF):
    a = (1 - mu_FM) + alpha * (1 - mu_MF)
    b = (1 - mu_FM) - alpha * (1 - mu_MF)
    delta = np.sqrt(a ** 2 - 4 * alpha * (1 - mu_FM - mu_MF))

    den = b + 2 * mu_FM + delta
    f = (b + delta) / den
    m = 2 * mu_FM / den

    return f - m
```
